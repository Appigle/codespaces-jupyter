# Response Time Analysis for E-commerce Application

This notebook analyzes the response times from five different user interactions with an e-commerce application. It calculates the `mean` and `variance` of the recorded response times to assess the application's performance under varying user loads.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


interaction_1 = [200, 220, 210, 190, 230, 240, 215, 205, 195, 225]
interaction_2 = [300, 320, 310, 290, 330, 340, 315, 305, 295, 325]
interaction_3 = [174, 200, 198, 195, 180, 191, 195, 198, 200, 192]
interaction_4 = [209, 221, 220, 220, 215, 208, 222, 210, 221, 230]
interaction_5 = [240, 252, 274, 255, 267, 239, 242, 236, 247, 257]

interactions = [
    np.array(interaction_1),
    np.array(interaction_2),
    np.array(interaction_3),
    np.array(interaction_4),
    np.array(interaction_5)
]

means = [np.mean(interaction) for interaction in interactions]
variances = [np.var(interaction) for interaction in interactions]
```

```python
data = {
    "Interaction": [f"Interaction {i+1}" for i in range(5)],
    "Mean Response Time (ms)": means,
    "Variance (ms^2)": variances
}
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8, 5))
plt.bar([f"Interaction {i+1}" for i in range(5)], means, color='skyblue')
plt.title("Mean Response Time per Interaction")
plt.xlabel("Interaction")
plt.ylabel("Mean Response Time (ms)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.bar([f"Interaction {i+1}" for i in range(5)], variances, color='salmon')
plt.title("Variance in Response Time per Interaction")
plt.xlabel("Interaction")
plt.ylabel("Variance (ms^2)")
plt.tight_layout()
plt.show()
```

## Decision and Implementation Report

After checking the response times, The Interactions 2 and 5 have much higher variance than the others. This means those two parts of the app dont always respond the same waysometimes theyre fast, and other times theyre slow. 

To improve this, we should check whats happening in the backend during these interactions. There might be inefficient code or unnecessary steps slowing things down. Also, its important to review the database queries being usedmaybe they need better indexing or could be simplified. Caching results might help, too.

Besides, we should monitor how much traffic each server is handling. If some servers are overloaded, using load balancing could help spread the work more evenly. For tasks that dont need to be instant, using background processes might make the app feel quicker.

Lastly, users feel better if they see a loading spinner or some kind of feedback instead of nothing happening. That way, they know the app is working. The app should feel smoother and more reliable for everyone using it.
